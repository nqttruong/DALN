```python
import pandas as pd
import os 
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = r"D:\CNTT\py_for_da\crawl_data_analyst\tiki"
os.listdir(path)
```

```python
frames = []
for file in os.listdir(path):
    if file.endswith('.csv'):
        filepath = path + '\\' + file
        df1 = pd.read_csv(filepath)
        frames.append(df1)
        result = pd.concat(frames)

df_all = result
df_all.to_csv("tk_7days_all.csv", index="False")
```

## 7. Top 10 mặt hàng

## 7.1. Theo doanh số cao nhất 

```python
top_10_items = df_all.groupby('item_list')['revenue'].sum().reset_index()
top_10_items = top_10_items.nlargest(10, 'revenue')

top_10_items
```

```python
# Lấy dữ liệu từ DataFrame
item_list = top_10_items['item_list']
revenue = top_10_items['revenue']

# Chia giá trị doanh thu cho 1 tỷ để hiển thị nghìn tỷ
revenue_thousand_billion = revenue / 1000000000

# Vẽ biểu đồ cột đứng
plt.figure(figsize=(10, 6))
bars = plt.barh(item_list, revenue_thousand_billion, color='skyblue')
plt.xlabel('Doanh thu (tỷ đồng)')
plt.ylabel('Loại sản phẩm')
plt.title('Top 10 loại sản phẩm có doanh thu cao nhất')

# Hiển thị số liệu trên mỗi cột
for bar, revenue in zip(bars, revenue_thousand_billion):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{revenue:.2f}', 
             va='center', ha='left')

plt.gca().invert_yaxis()  # Đảo ngược trục y để loại sản phẩm có doanh thu cao nhất ở trên cùng
plt.show()
```

## 7.2 Theo sản phẩm bán chạy

```python
top_10_items = df_all.groupby('item_list')['sales'].sum().reset_index()
top_10_items = top_10_items.nlargest(10, 'sales')
top_10_items
```

```python
item_list = top_10_items['item_list']
sales = top_10_items['sales']

# Chia giá trị doanh thu cho 1 tỷ để hiển thị nghìn tỷ
sales_thousand_billion = sales / 1000

# Vẽ biểu đồ cột đứng
plt.figure(figsize=(10, 6))
bars = plt.barh(item_list, sales_thousand_billion, color='skyblue')
plt.xlabel('Sản phẩm (nghìn sản phẩm)')
plt.ylabel('Loại sản phẩm')
plt.title('Top 10 loại sản phẩm bán chạy nhất')

# Hiển thị số liệu trên mỗi cột
for bar, sales in zip(bars, sales_thousand_billion):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{sales:.2f}', 
             va='center', ha='left')

plt.gca().invert_yaxis()  # Đảo ngược trục y để loại sản phẩm có doanh thu cao nhất ở trên cùng
plt.show()
```
