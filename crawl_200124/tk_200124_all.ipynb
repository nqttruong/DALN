```python
import pandas as pd
import os 
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = r"D:\CNTT\py_for_da\crawl_data_analyst\tiki\crawl_200124"
os.listdir(path)
```

```python
frames = []
for file in os.listdir(path):
    if file.endswith('.csv'):
        filepath = path + '\\' + file
        df1 = pd.read_csv(filepath)
        frames.append(df1)
        result = pd.concat(frames)

df_all = result
df_all.to_csv("tk_200124_all.csv", index="False")
```
